```python
data.drop(['modelyr','cylinder','org','carname'],axis=1,inplace=True)
```

```python
data.columns
```

```python
y=data.mpg
print(y)
```

```python
x=data.drop('mpg',axis=1)
```

```python
x.head()
```

```python
x=sm.add_constant(x)
```

```python
train_y=y.iloc[0:319]
test_y=y.iloc[319:]
```

```python
train_x=x.iloc[0:319]
test_x=x.iloc[319:]
```

```python
train_x.shape
```

```python
model=sm.OLS(train_y,train_x).fit() 
```

```python
predicted_y=model.predict(test_x)
```

```python
model.summary()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=train_x.hpower,y=train_y)
sns.lineplot(x=test_x.hpower,y=test_y,color='green')
sns.lineplot(x=test_x.hpower,y=predicted_y,color='red')
```
